### Ajout des specifs
  

```python
data = pandas.merge(data_MC, data_specif1,  
                   on='token',  
                   how='inner')
data = pandas.merge(data, data_specif2,  
                   on='token',  
                   how='inner')
print(data)
specific_lim_x = data_specif1['specif'][int(len(data_specif1)/4)]
specific_lim_y = data_specif2['specif'][int(len(data_specif2)/4)]
print(specific_lim_x, specific_lim_y)
```

```python
tres_specific_historique = data.loc[data['specif_x'] <= specific_lim_x]
print(tres_specific_historique)
tres_specific_news = data.loc[data['specif_y'] <= specific_lim_y]
print(specific_lim_x, specific_lim_y)
print(tres_specific_news)
```

```python
import jieba

with open("../../corpus/corpus_discours/paragraph/2.txt", 'r') as f:
    paragraphs = f.readlines()

for paragraph in paragraphs:
    #words = jieba.cut_for_search(paragraph)
    words = list(jieba.cut(paragraph))
    
    score_historique = 0
    n_historique = 0
    score_news = 0
    n_news = 0
    mots_communs = 0
    for word in words:
        specif = data.loc[data['token'] == word, 'specif_y'] # specif_x = historique, specif_y = news
        #print(type(specif))
        specif = list(specif)
        #print(len(specif), specif, word)
        if len(specif) > 0 and specif[0] < -0:
            print("\x1b[94m" + str(specif[0]) + "\x1b[0m", end = "/")
            score_news += specif[0]
            n_news += 1
            
        specif = data.loc[data['token'] == word, 'specif_x'] # specif_x = historique, specif_y = news
        #print(type(specif))
        specif = list(specif)
        #print(len(specif), specif, word)
        if len(specif) > 0 and specif[0] < -0:
            print("\x1b[92m" + str(specif[0]) + "\x1b[0m", end = "/")
            score_historique += specif[0]
            n_historique += 1
            
        #print(word)
        if str(word) in data_MC['token'].values:
            #html_print(cstr(word, 'red'))
            #print(word, color='red')
            print("\x1b[31m" + word + "\x1b[0m", end= " ")
            mots_communs += 1
        else:
            print(word, end=' ')
    score_historique /= len(words)
    score_news /= len(words)
    mots_communs /= len(words)
    n_historique /= len(words)
    n_news /= len(words)
    print("[ historique:" + str(score_historique) + " news:" + str(score_news) + " communs: " + str(mots_communs) + "]")
    print(" [ n historique: " + str(n_historique) + " n_news: " + str(n_news) + "]")
    print(" [ " + str( - n_historique * score_historique) + ", " + str(- n_news * score_news) + " ] ")
```
